Demo for ensure_annotations:

The type hints (x:int, y:int -> int) are purely for documentation and static analysis tools (like MyPy) in standard Python. They do not enforce type checking at runtime by default.

```python
# Default python syntax
def get_prodict(x:int, y:int) -> int:
    return x*y

print(get_prodict(x=2, y=4), type(get_prodict(x=2, y=4)))
print(get_prodict(x=2, y="4"), type(get_prodict(x=2, y="4")))
print("================================")

# With ensure_annotations
from ensure import ensure_annotations

@ensure_annotations
def get_prodict(x:int, y:int) -> int:
    return x*y

print(get_prodict(x=2, y=4), type(get_prodict(x=2, y=4)))
print(get_prodict(x=2, y="4"))
```
